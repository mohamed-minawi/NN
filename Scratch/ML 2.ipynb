```python
import Load_Data as ld
import LayerFunc as lf
import NN as Neural
import NN_Trainer as nnt
import numpy as np
import matplotlib.pyplot as plt
```

```python
arrays = ld.get_CIFAR10_data()
```

```python
lr = 0.0001
ws = 0.01
reg = 0.01

np.random.seed(7)
```

```python
NNmodel = Neural.NN(features = 3072, classes = 10, hid_layers = [1000, 600, 200], reg = reg, weight = ws )

trainer = nnt.NN_Trainer(model = NNmodel, data= arrays ,lr = lr, batch_size = 500, num_epochs = 50,
                        print_every = 50, verbose = 1, optim_config = {'learning_rate': lr})
```

```python
trainer.train()
```

```python
y_test_pred = np.argmax(NNmodel.loss(arrays['X_test']), axis =1 )
print('ACCR: ', (y_test_pred == arrays["y_test"]).mean())

for i in range(10):
    print(i,  np.sum( np.logical_and(y_test_pred == arrays["y_test"], arrays["y_test"] == i)) / float(np.sum(arrays["y_test"] == i)) )
```

```python
plt.plot(trainer.val_acc_history, 'o')
plt.title('Validation Accuracy vs Epochs')
plt.xlabel('Iteration')
plt.ylabel('Val Acc')
plt.show()
```

```python
plt.plot(trainer.train_acc_history, 'o')
plt.title('Training Accuracy vs Epochs')
plt.xlabel('Iteration')
plt.ylabel('Train Acc')
plt.show()
```

```python
plt.plot(trainer.loss_history, 'o')
plt.title('Training Loss vs Epochs')
plt.xlabel('Iteration')
plt.ylabel('Train Loss')
plt.show()
```

```python
plt.plot(trainer.val_loss, 'o')
plt.title('Validation Loss vs Epochs')
plt.xlabel('Iteration')
plt.ylabel('Val Loss')
plt.show()
```
